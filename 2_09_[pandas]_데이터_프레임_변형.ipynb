```python
import pandas as pd

filepath="/content/drive/MyDrive/ICTCOG 인공지능과정/Data/korean-idol.csv"
df = pd.read_csv(filepath)
df.head()
```

# 피벗 테이블(pivot table)
* 데이터 프레임의 데이터를 인덱스로 사용하기 위한 테이블
* 항목 데이터를 컬럼이나 행의 인덱스로 만들기 위함
  * 내가 원하는 항목 별로 행 / 열 형태로 집계가 가능
   * 소속사별 혈액형 마다의 브랜드 평판 지수
  * 집계는 보통 평균으로 낸다.
* 왼쪽에 나타나는 인덱스를 **행 인덱스**, 상단에 나타나는 인덱스를 **열 인덱스**라고 한다.

```python
pd.pivot_table(df,
               index="소속사", # 행 인덱스 설정
               columns="성별") # 열 인덱스 설정

# 평균을 계산하는 값은 수치 데이터 타입이 계산된다.
```

```python
pd.pivot_table(df,
               index="그룹",
               columns="혈액형",
               values="키") # value : 조회하고 싶은 값을 집계해서 보여줘라!
```

```python
import numpy as np

# 직접 집계 함수를 넣어 줄 수도 있다.
pd.pivot_table(df,
               index="그룹",
               columns="혈액형",
               values="브랜드평판지수",
               aggfunc=[np.sum, np.mean])
```

```python
pd.pivot_table(df,
               index="혈액형",
               columns="그룹",
               values="브랜드평판지수",
               aggfunc=[np.sum, np.mean])
```

# 그룹핑
* groupby 함수를 활용
* 데이터를 **그룹으로 묶어서 분석할 때 유용**
* **~ 별로 집계**

```python
# 소속사별 무언가(집계)를 해보고 싶어요.
df.groupby("소속사") # 이 상태에서 필요한 집계를 수행
```

집계의 종류
* count() - 개수
* sum() - 합계
* mean() - 평균
* var() - 분산
* std() - 표준편차
* min() - 최소값
* max() - 최대값

```python
df.groupby("소속사").count()
```

산술 통계(평균, 분산, 표준편차)는 자동으로 산술 통계가 가능한 컬럼만 집계가 됩니다.

```python
df.groupby("그룹").mean()
```

```python
df.groupby("성별").sum()
```

```python
# 혈액혈병 키의 평균
df.groupby("혈액형")['키'].mean()
```
