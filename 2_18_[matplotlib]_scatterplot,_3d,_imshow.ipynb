# 1. Scatterplot

## 1 - 1. x, y, colors, area 설정
* colors는 임의 값을 color 값으로 변환
* area 점의 넓이 (크기)를 나타냄. 값이 커지면 넓이도 커짐

```python
x = np.random.rand(50)
y = np.random.rand(50) # 랜덤하게 50개를 뽑은 array

colors = np.arange(50) # 0~49까지 순서가 있는 array
area = x * y * 1000
```

```python
plt.scatter(x,y, 
            s = area, # 점의 크기 설정(s)
            c = colors) # 점의 색상 설정(c)
plt.show()
```

## 1 - 2. cmap과 alpha
* cmap에 컬러를 지정하면, 컬러 값을 모두 같게 가져갈 수도 있다.
    * 색상 단일 설정
* alpha값은 투명도를 나타낸다. 0~1 사이의 값을 지정. 0에 가까울수록 투명

```python
# canvas 설정
plt.figure(figsize=(12,6)) # figsize : 캔버스의 크기 설정

plt.subplot(1,3,1)
plt.scatter(x,y, s=area, c = y, alpha = 0.1)
plt.title('alpha = 0.1')
plt.colorbar()

plt.subplot(1,3,2)
plt.scatter(x,y, s=area, cmap = 'blue', alpha = 0.8)
plt.title('alpha = 0.3')
plt.colorbar()

plt.subplot(1,3,3)
plt.scatter(x,y, s=area, c = y * 100, alpha = 1.0)
plt.title('alpha = 1.0')
plt.colorbar()


plt.show()
```

# 2. Barplot, Barhplot

## 2 - 1. 기본 Barplot 그리기

```python
x = ['수학', '과학', '영어', '컴퓨터', '미술', '체육']
y = [66,80,85,50,80,90]

plt.bar(x,y, align='center', alpha= 0.7, color = 'red')

plt.xticks(x) # x축에 표시할 값들

plt.xlabel('과목', fontsize= 15)
plt.ylabel('점수', fontsize= 15)
plt.title('성적', fontsize = 20)
plt.show()
```

## 2 - 2. 기본 Barplot 그리기
* barh 함수에서는 **xticks로 설정**했던 부분을 **yticks**로 변경

```python
x = ['수학', '과학', '영어', '컴퓨터', '미술', '체육']
y = [66,80,85,50,80,90]

plt.barh(x,y, align='center', alpha= 0.7, color = 'green')

plt.yticks(x) # y축에 표시할 값들

plt.xlabel('과목', fontsize= 15)
plt.ylabel('점수', fontsize= 15)
plt.title('성적', fontsize = 20)
plt.show()
```

### Barplot에서 비교 그래프 그리기

```python
x_label = ['수학', '과학', '영어', '컴퓨터', '미술', '체육']
x = np.arange(len(x_label))

y_1 = [66,80,85,50,80,90]
y_2 = [86,75,91,43,77,80]

# 넓이 지정
width = 0.35

# subplots 생성
fig, axes = plt.subplots() #매개 변수 없으면 기본 figure 1 × 1

axes.bar(x - width / 2, y_1, width, align = 'center', alpha=0.5)
axes.bar(x + width / 2, y_2, width, align = 'center', alpha=0.8)

plt.xticks(x)
axes.set_xticklabels(x_label)

plt.ylabel('점수')
plt.xlabel('과목')
plt.legend(['학생A', '학생B'])

plt.show()
```

# 3. Line Plot

## 3 - 1. 기본 lineplot 그리기

```python
x = np.arange(0,10,0.1)
y = 1 + np.sin(x) #각각의 x값에 대한 삼각함수 sin값 만들기

plt.plot(x,y)

plt.xlabel('x value')
plt.ylabel('y value')
plt.title('sin graph')

plt.grid()

plt.show()
```

## 3 - 2. 2개 이상의 그래프 그리기

```python
x = np.arange(0,10,0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

# 하나의 캔버스에 2개 그래프 그리기
plt.plot(x,y_1,color = 'blue')
plt.plot(x,y_2,color = 'red')

plt.xlabel('x value')
plt.ylabel('y value')
plt.title('sin / cos graph')

plt.grid()
plt.legend(['1+sin', '1+cos'])

plt.show()
```

## 3-3. 마커 스타일링

```python
x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

# 하나의 캔버스에 2개의 그래프 그리기 + 마커 추가
plt.plot(x, y_1, color='blue', alpha=0.3, marker='o')
plt.plot(x, y_2, color='red', alpha=0.7, marker='+')

plt.xlabel("x value (time)")
plt.ylabel("y value (amplitude)")
plt.title("sin / cos graph")

plt.grid()
plt.legend(["1+sin", "1+cos"])

plt.show()
```

## 3-4. 라인 스타일 변경하기

```python
x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

# 하나의 캔버스에 2개의 그래프 그리기 + 마커 추가 + 라인 스타일 변경
plt.plot(x, y_1, color='blue', alpha=0.3, marker='o', linestyle=':')
plt.plot(x, y_2, color='red', alpha=0.7, marker='+', linestyle='-.')

plt.xlabel("x value (time)")
plt.ylabel("y value (amplitude)")
plt.title("sin / cos graph")

plt.grid()
plt.legend(["1+sin", "1+cos"])

plt.show()
```

# 5. Histogram

```python
N = 100000 # 데이터의 개수

# 중요한 변수
bins = 30 # 30개씩 끊겠다는 의미 - 범위

x = np.random.randn(N)

plt.hist(x, bins=bins)
plt.show()
```

* sharey : sharey가 True면, y축 값을 다중 그래프가 공유
* tight_layout : graph의 패딩을 자동으로 조절해서 fit한 그래프를 생성

```python
fig, axes = plt.subplots(1,3,
                         sharey = True, # y축을 모든 그래프가 공유
                         tight_layout = True # 그래프들이 여백 없이 들어감
                         )
# 크기 조절
fig.set_size_inches(12,5)

axes[0].hist(x, bins=bins)
axes[1].hist(x, bins = bins*2)
axes[2].hist(x, bins = bins*4)
plt.show()
```

## 5 - 3. Y축에 밀집도(Density) 표기

```python
N = 100000
bins = 30

x = np.random.randn(N)

fig, axes = plt.subplots(1, 2, tight_layout=True)
fig.set_size_inches(9,3)

# density = True 값을 통하여 Y축에 Density를 표기할 수 있다.
axes[0].hist(x, bins = bins, density = True, cumulative = True) # cumulative : 갯수 누적
axes[1].hist(x, bins = bins, density = True)
plt.show()
```

# 6. Pie Chart
**pie chart 옵션**
* explode : 파이에서 툭 튀어져 나온 비율
* autopct : 퍼센트 자동 표기
* shadow : 그림자 표시
* startangle : 파이를 그리기 시작할 각도

text, autotexts 인자를 파이차트를 그리면 리턴 받음.

* texts : label에 대한 텍스트 효과
* autotexts : 파이 위에 그려지는 텍스트 효과를 다룰 때 사용

```python
labels = ['A', 'B', 'C', 'D', 'E', 'F']
sizes = [20.4, 15.8, 10.5, 9, 7.6, 36.7]

explode = (0.3, 0, 0, 0, 0, 0)

patches, texts, autotexts = plt.pie(
    sizes,
    explode = explode,
    labels = labels,
    autopct = '%1.1f%%',
    shadow = True,
    startangle = 90
)

# label 텍스트 스타일링
for t in texts:
    t.set_fontsize(12)
    t.set_color('blue')

# pie 위의 텍스트에 대한 스타일링

for t in autotexts:
    t.set_fontsize(12)
    t.set_color('white')
    
plt.title('파이차트 그려보기!')

plt.show()
```

# 7. Box Plot
* 데이터의 최대, 최소, 1Q, 3Q 지점 등 통계적인 부분들을 손쉽게 시각화
* 이상치도 빠르게 판단이 가능

```python
# 기본 데이터
spread = np.random.rand(50) * 100 # 최소 0, 최대 100
center = np.ones(25) * 50

# high : 최대, low : 최소
flier_high = np.random.rand(10) * 100 + 100 # 최소 100 ~ 최대 200
flier_low  = np.random.rand(10) * -100 # 최소 -100 ~ 최대 0

# 위에서 만든 4가지 데이터를 모두 합치기
data = np.concatenate([spread, center, flier_high, flier_low])
data
```

# 7-1. 기본 Boxplot 생성

```python
# 음수 나오게 하는 설정
plt.rcParams["axes.unicode_minus"] = False

plt.boxplot(data)
plt.show()
```

## 7-2. Box plot 축 변경하기
`vert=False` 옵션을 이용해 표시하고자 하는 축을 변경할 수 있다.

```python
plt.boxplot(data, vert=False)
plt.show()
```

## 7-3.outlier marker 심볼, 컬러 변경

```python
# 옵션용 dict 생성
outlier_marker = dict(markerfacecolor='r', marker='D')
outlier_marker
```

```python
plt.boxplot(data, flierprops=outlier_marker) # flierprops : outlier 표시용 점들
plt.show()
```

# 8. 3D 그래프 그리기

## 8-1. 3D 그래프의 캔버스 확인하기

```python
from mpl_toolkits import mplot3d

fig = plt.figure()

# 시각화 축을 3D로 지정
ax = plt.axes(projection='3d')
```

## 8-2. 3d plot 그리기

```python
# projection='3d' 설정
ax = plt.axes(projection='3d')

# x, y, z 데이터가 필요
# 단순한 numpy 배열이면 된다.
z = np.linspace(0, 15, 1000)
x = np.sin(z)
y = np.cos(z)

ax.plot(x, y, z, 'gray')
plt.show()
```

```python
ax = plt.axes(projection='3d')

sample_size = 100

# cumsum : 누적합
x = np.cumsum(np.random.normal(0, 1, sample_size)) # ex ) [0, 1, 2] -> [0, 1, 3]
y = np.cumsum(np.random.normal(0, 1, sample_size)) # ex ) [4, 5, 3] -> [4, 9, 12]
z = np.cumsum(np.random.normal(0, 1, sample_size))

# plt.plot(x, y, z, alpha=0.6, marker='o')
ax.plot3D(x, y, z, alpha=0.6, marker='o')
plt.show()
```

## 8-3. 3d-scatter 그리기

```python
# fig 사이즈를 조절해서 그래프 크기 변경하기
fig = plt.figure(figsize=(10, 5)) # canvas를 설정
ax = fig.add_subplot(111, projection='3d') # Axes3D object


sample_size = 500

x = np.cumsum(np.random.normal(0, 5, sample_size))
y = np.cumsum(np.random.normal(0, 5, sample_size))
z = np.cumsum(np.random.normal(0, 5, sample_size))

ax.scatter(x, y, z, c=z, s=20, alpha=0.5, cmap='Greens')

plt.show()
```

## 8-4 countour3D 그리기 (등고선)


```python
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

xx, yy = np.meshgrid(x, y)
z = np.sin(np.sqrt(xx**2 + yy**2)) # 높이

fig = plt.figure(figsize=(12, 6)) # canvas설정
ax  = plt.axes(projection='3d')   # 그래프 관련 설정

ax.contour3D(xx, yy, z, 20, cmap='Reds')
plt.show()
```

# 9. imshow ★★★★★
이미지 데이터를 시각화 - 이미지 (2차원 배열 )를 사용

```python
from sklearn.datasets import load_digits # MNIST 불러오기

digits = load_digits() # MNISt 데이터셋 불러오기

X = digits.images[:10] # 앞에서 부터 10개의 이미지 불러오기

X[0] # 첫 번째 이미지에 대한 정보를 출력
```

```python
# 첫 번째 이미지 출력해 보기
plt.imshow(X[0], cmap='Blues')
plt.show()
```
