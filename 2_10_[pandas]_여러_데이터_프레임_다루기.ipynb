# DataFrame 병합 - merge
* concat과 merge는 단순하게 합쳐지는 목적 또는 특정 기준(index)으로 합치느냐에 따라서 용도가 달라집니다.
* concat : row나 column 기준으로 단순하게 이어 붙히기
* merge : 특정 고유한 키(unique id)값을 기준으로 병합하기


```python
df1.head()
```

```python
df2.head()
```

df1과 df2는 '이름'이라는 column이 겹친다.

우리는 '이름'이라는 기준으로 df1, df2를 병합할 것임!

**pd.merge(left, right, on='기준 컬럼', how='left')**
* left와 right는 병합할 두 DataFrame을 대입
* on에는 병합의 기준이 되는 컬럼의 이름
* how에는 `'left', 'right', 'inner', 'outer'`라는 4가지 병합 방식을 지원

### left, right 방식 살펴보기

```python
df2_right = df2.drop([1, 3, 5, 7, 9], axis=0)
df2_right.shape
```

```python
df2_right = df2_right.reset_index(drop=True)
df2_right
```

합치기 전에 shape 확인하기

```python
df1.shape, df2_right.shape
```

```python
pd.concat([df1, df2_right], axis=1)
```

* 'left' 옵션을 부여하면, left DataFrame에 키 값이 존재하면 해당 데이터를 유지
  * right DataFrame에 데이터가 없으면 NaN값이 들어간다.

```python
pd.merge(df1, df2_right, on='이름', how='left')
```

반대로 'right'를 기준으로 두면??
* 오른쪽 데이터 프레임을 기준으로 합치게 된다.
* 왼쪽 데이터 프레임이 데이터의 개수가 더 많죠? (15개), 오른쪽은 10개
* 오른쪽이 기준이 되기 때문에 왼쪽 데이터 프레임의 데이터는? 삭제가 된다.

```python
pd.merge(df1, df2_right, on='이름', how='right')
```

## inner, outer 방식
* inner : 두 데이터 프레임에 모두 키값이 존재하는 경우만 병합
* outer : 하나의 데이터 프레임에 키값이 존재하는 경우 모두 병합합니다.
* outer 방식에서 없는 값은 NaN으로 대입

```python
pd.merge(df1, df2_right, on='이름', how='inner')
```

```python
pd.merge(df1, df2_right, on='이름', how='outer')
```

# [번외] 컬럼의 이름은 다르지만, 동일한 성질의 데이터인 경우는??

df1에는 "이름", df2에는 "성함"

```python
df3 = df2.rename(columns={"이름":"성함"})
df3
```

df1은 이름, df3는 성함으로 되어 있기 때문에 **기준이 되는 컬럼을 지정할 수 없다.** ( on을 사용할 수 없다. )

```python
pd.merge(df1, df3, 
         left_on='이름', # 왼쪽 데이터 프레임의 기준
         right_on='성함' # 오른쪽 데이터 프레임의 기준
         )
```
